```python
import numpy as np
import pandas as pd
df=pd.read_csv('winequality.csv')
df
```

```python
df.info()
```

```python
df[df['fixed acidity'].isna()]
```

```python
#fixed acidity sÃ¼tununu her ÅŸarap tipinin ortalamasÄ±na gÃ¶re nan deÄŸerleri dolduralÄ±m
df['fixed acidity'] = df.groupby('type')['fixed acidity'].apply(lambda x: x.fillna(x.mean()))
```

```python
df[df['volatile acidity'].isna()]
```

```python
#volatile acidity sÃ¼tununu her ÅŸarap tipinin ortalamasÄ±na gÃ¶re nan deÄŸerleri dolduralÄ±m
df['volatile acidity'] = df.groupby('type')['volatile acidity'].apply(lambda x: x.fillna(x.mean()))
```

```python
df[df['citric acid'].isna()]
```

```python
#citric acid sÃ¼tununu her ÅŸarap tipinin ortalamasÄ±na gÃ¶re nan deÄŸerleri dolduralÄ±m
df['citric acid'] = df.groupby('type')['citric acid'].apply(lambda x: x.fillna(x.mean()))
```

```python
df[df['residual sugar'].isna()]
```

```python
#residual sugar sÃ¼tununu her ÅŸarap tipinin ortalamasÄ±na gÃ¶re nan deÄŸerleri dolduralÄ±m
df['residual sugar']=df['residual sugar'].fillna(df[df['type']=='white']['residual sugar'].mean())
```

```python
df[df['chlorides'].isna()]
```

```python
#chlorides sÃ¼tununu her ÅŸarap tipinin ortalamasÄ±na gÃ¶re nan deÄŸerleri dolduralÄ±m
df['chlorides']=df['chlorides'].fillna(df[df['type']=='white']['chlorides'].mean())
```

```python
df[df['pH'].isna()]
```

```python
#pH sÃ¼tununu her ÅŸarap tipinin ortalamasÄ±na gÃ¶re nan deÄŸerleri dolduralÄ±m
df['pH'] = df.groupby('type')['pH'].apply(lambda x: x.fillna(x.mean()))
```

```python
df[df['sulphates'].isna()]
```

```python
#sulphates sÃ¼tununu her ÅŸarap tipinin ortalamasÄ±na gÃ¶re nan deÄŸerleri dolduralÄ±m
df['sulphates'] = df.groupby('type')['sulphates'].apply(lambda x: x.fillna(x.mean()))
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['type'] = le.fit_transform(df['type'])
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
independent_variables=df.drop(['quality'], axis=1)
predict_value=df['quality']
X_train, X_test, y_train, y_test = train_test_split(independent_variables, predict_value, test_size=0.20, random_state=0)
logistic_classifier= LogisticRegression(random_state=0)  
logistic_classifier.fit(X_train, y_train)
y2=logistic_classifier.predict(X_train)
mse = mean_squared_error(y_train, y2)
mse
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()  
lr.fit(X_train, y_train)
y2=lr.predict(X_train)
mse = mean_squared_error(y_train, y2)
mse
```

```python
import xgboost as xgb
xgbr = xgb.XGBRegressor(verbosity=0) 
xgbr.fit(X_train, y_train)
y2=xgbr.predict(X_train)
mse = mean_squared_error(y_train, y2)
mse
```
